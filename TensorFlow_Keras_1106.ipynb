```python
import numpy as np
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import SGD
```

```python
## load data
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("MNIST_data/", one_hot = True)

train_xdata = np.array([np.reshape(x, (28*28)) for x in mnist.train.images])
test_xdata = np.array([np.reshape(x, (28*28)) for x in mnist.test.images])
train_labels = mnist.train.labels
test_labels = mnist.test.labels
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Activation
```

```python
x_in = train_xdata.shape[1]         #input=784, output=50
train_xdata.shape
```

```python
n_hidden=50
n_out = train_labels.shape[1]
```

```python
## hidden layer
model = Sequential()
model.add(Dense(n_hidden, input_dim=x_in))
model.add(Activation("relu"))
```

```python
## output layer
model.add(Dense(n_out, input_dim=n_hidden))
model.add(Activation("softmax"))             # 마지막에는 cross entropy 구하기 위한 softmax

model.compile(loss="categorical_crossentropy", optimizer=SGD(lr=0.01), metrics=['accuracy'])
```

```python
epochs=100
batch_size = 100
model.fit(train_xdata, train_labels, epochs=epochs, batch_size=batch_size)
```

```python
loss_and_matrics=model.evaluate(test_xdata, test_labels)
```

```python
print(loss_and_matrics)
```

```python
n_hidden=50
n_out = train_labels.shape[1]

## hidden layer
model = Sequential()
model.add(Dense(n_hidden, input_dim=x_in))
model.add(Activation("relu"))

model.add(Dense(n_hidden+10, input_dim=n_hidden))         # n_hidden=50 이 input
model.add(Activation("relu"))

## output layer
model.add(Dense(n_out, input_dim=n_hidden+10))
model.add(Activation("softmax"))             # 마지막에는 cross entropy 구하기 위한 softmax

model.compile(loss="categorical_crossentropy", optimizer=SGD(lr=0.01), metrics=['accuracy'])

epochs=100
batch_size = 100
model.fit(train_xdata, train_labels, epochs=epochs, batch_size=batch_size)
```

```python
loss_and_matrics=model.evaluate(test_xdata, test_labels)
print(loss_and_matrics)
```
