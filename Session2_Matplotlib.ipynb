```python
## Contour plot
```

```python
def f(x,y):
    return x**2 + y**2

x=np.linspace(-2,2, 20)
y=np.linspace(-4,4, 30)
X,Y = np.meshgrid(x,y)
```

```python
[X.shape, X.ndim, X.size]
```

```python
[Y.shape, Y.ndim, Y.size]
```

```python
Z = f(X,Y)
```

```python
plt.contour(X,Y,Z,20,  cmap="RdGy")
plt.colorbar()
```

```python
plt.contourf(X,Y,Z,20,  cmap="RdGy")
plt.colorbar()
```

```python
plt.imshow(Z, extent=[-2,2, -4, 4], origin="lower", cmap="RdGy")
plt.colorbar()
plt.axis(aspect="image")
```

```python
# Two plots at the same time
contours = plt.contour(X,Y,Z,3, colors="k")
plt.clabel(contours, inline=True, fontsize=8)
plt.imshow(Z, extent=[-2, 2, -4, 4], origin="lower", cmap="RdGy", alpha=0.5)
plt.colorbar()
```

## 3-D plot

```python
from mpl_toolkits import mplot3d
fig = plt.figure()
ax=plt.axes(projection="3d")
```

```python
ax=plt.axes(projection="3d")
zline = np.linspace(0,15, 100)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, "gray")
```

```python
ax=plt.axes(projection="3d")
ax.scatter3D(xline, yline, zline, c=zline, cmap="Greens")
```

```python
def f(x,y):
    return x**3 + y**2

x= np.linspace(-1, 1, 30)
y= np.linspace(-1, 1, 30)
X,Y = np.meshgrid(x,y)
Z=f(X,Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.contour3D(X,Y,Z, 50, cmap="binary")
```
