# contour plot

```python
def f(x,y):
    Z = x**2 + y**2
    return Z
```

```python
x = np.linspace(-2,2,20)           # -2부터 2까지 20개의 점
y = np.linspace(-4,4,30)d

# broadcasting과 비슷.
X,Y = np.meshgrid(x,y)
```

```python
X.shape
Y.shape
```

```python
Z = f(X,Y)
```

```python
Z.shape
```

```python
plt.contour(X,Y,Z, 20, cmap='RdGy')             # 빨간색이면 높이가 높고, 검정색이면 높이가 낮음.
plt.colorbar()
```

```python
plt.contourf(X,Y,Z, 20, cmap='RdGy')             # contourf 로 쓰면 색을 칠해준다.
plt.colorbar()
```

```python
plt.contourf(X,Y,Z, 100, cmap='RdGy')             # 20개가 아니라 100개로 나눠주므로 더 세밀하게 표현됨.
plt.colorbar()
```

```python
plt.imshow(Z)               # 높은 것을 하얀색, 낮은 것을 검정색으로 표현해준다.
```

```python
plt.imshow(Z, extent=[-2,2,-4,4], origin = 'lower', cmap='RdGy')        # x:(-2,2), y:(-4,4)
plt.colorbar()
plt.axis(aspect='image')         # 축 범위 나타내줌.
```

# two plots at the same time

```python
contours = plt.contour(X,Y,Z,3)
plt.clabel(contours, inline=True, fontsize=8)                # 등고선 위에 글씨 쓰기.
plt.imshow(Z, extent=[-2,2,-4,4], origin='lower')            # origin : 방향
plt.colorbar()
```

# Multiple Subplot
# plt.axes : Subplots by Hand

```python
ax1 = plt.axes()      #standard axes
ax2 = plt.axes([0.65,0.65,0.2,0.2])      #[bottom, left, width, height]

# (0.65, 0.65) 위치에 가로축과 세로축 각각의 20% subplot 그림.
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.5,0.8,0.4], xticklabels=[], ylim=(-1.2,1.2))
ax2 = fig.add_axes([0.1,0.1,0.8,0.4], ylim=(-1.2,1.2))

x = np.linspace(0,10)
ax1.plot(np.sin(x))
ax2.plot(np.cos(x))
```

# plt.subplot : Simple Grids of Subplots

```python
# grid 안에 여러개의 single plot 그리기.
for i in range(1,7):
    plt.subplot(2,3,i)
    plt.text(0.5, 0.5, str((2,3,i)), fontsize=18, ha='center')
```

```python
# plt.subplot() with adjusted margins
fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=0.4)
for i in range(1,7):
    ax = fig.add_subplot(2,3,i)
    ax.text(0.5, 0.5, str((2,3,i)), fontsize=18, ha='center')
```

```python
# 10x10 plot 생성.
# figsize : 가로세로 크기 10, subplot_kw : x,y축 범례 없애기
# gridspec_kw : plot사이 간격
fig, axes=plt.subplots(10,10, figsize=(10,10), 
                       subplot_kw={'xticks':[],'yticks':[]},
                      gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i, ax in enumerate(axes.flat):
    ax.imshow(digits.images[i], cmap='binary', interpolation='nearest')
    ax.text(0.05,0.05,str(digits.target[i]), transform=ax.transAxes,
           color='green')
```

# Three-Dimensional Plotting in Matplotlib

```python
from mpl_toolkits import mplot3d
fig = plt.figure()
ax = plt.axes(projection = '3d')
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

# data for three-dimensional line
zline = np.linspace(0,15,1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')

# data for three-dimensional scattered points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata,ydata,zdata, c=zdata, cmap='Greens');
```

```python
def f(x,y):
    return np.sin(np.sqrt(x**2 + y**2))
x = np.linspace(-6,6,30)
y = np.linspace(-6,6,30)
X,Y = np.meshgrid(x,y)
Z = f(X, Y)
```

```python
Z.shape
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X,Y,Z,50, cmap='binary')
ax.set_xlabel('x'); ax.set_ylabel('y'); ax.set_zlabel('z');
```

```python
ax.view_init(60,35)     #view 각도 변화.

# 60-degrees above the x-y plane, rotated 35-degrees contour-clockwise about the fig
fig
```
